```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 120)

raw_path =  r'C:\Users\AADITYA\Desktop\Supply Chain Risk Detection\data\raw'

datasets = {
    "supply_chain_resilience": os.path.join(raw_path, "supply_chain_resilience_dataset.csv"),
    "us_supply_chain_risk": os.path.join(raw_path, "data.csv"),
    "cargo2000": os.path.join(raw_path, "c2k_data_comma.csv"),
    "smart_logistics": os.path.join(raw_path, "smart_logistics_dataset.csv"),
    "dynamic_logistics": os.path.join(raw_path, "dynamic_supply_chain_logistics_dataset.csv"),
}

dfs = {}

for name, path in datasets.items():
    try:
        dfs[name] = pd.read_csv(path)
        print(f"✅ Loaded {name} → {dfs[name].shape[0]} rows, {dfs[name].shape[1]} cols")
    except Exception as e:
        print(f"❌ Failed to load {name}: {e}")
```

```python
for name, df in dfs.items():
    print(f"\n=== {name.upper()} ===")
    display(df.head(5))
    display(df.info())
    print("\nMissing Values.")
    display(df.isna().sum())
```

```python
# Example: If delay or risk columns exist, visualize them
for name, df in dfs.items():
    print(f"\n=== {name.upper()} ===")
    for col in df.columns:
        if "delay" in col.lower() or "risk" in col.lower():
            plt.figure(figsize=(6,4))
            sns.histplot(df[col].dropna(), bins=30, kde=True)
            plt.title(f"{name} - {col} distribution")
            plt.show()
```

```python
for name, path in datasets.items():
    df = pd.read_csv(path, low_memory=False)
    print(f"\n🔍 Missing Value Analysis — {name}")
    
    # Replace '?' with NaN temporarily for counts
    df.replace("?", np.nan, inplace=True)
    
    missing = df.isna().sum()
    missing = missing[missing > 0].sort_values(ascending=False)
    
    if not missing.empty:
        display(missing.head(15))  # show top 15 cols with missing
    else:
        print("No missing values detected.")

```

```python
for name, path in datasets.items():
    df = pd.read_csv(path, low_memory=False)
    print(f"\n📊 Summary Stats — {name}")
    display(df.describe(include="all").transpose().head(10))
```

```python
for name, path in datasets.items():
    df = pd.read_csv(path, low_memory=False)
    df.replace("?", np.nan, inplace=True)
    
    print(f"\n📈 Distribution Plots — {name}")
    
    # Numerical distribution
    num_cols = df.select_dtypes(include=[np.number]).columns[:3]  # first 3 only
    for col in num_cols:
        plt.figure(figsize=(6,3))
        sns.histplot(df[col].dropna(), bins=30, kde=True)
        plt.title(f"{name} — {col}")
        plt.show()
    
    # Categorical frequency
    cat_cols = df.select_dtypes(include=["object"]).columns[:2]  # first 2 only
    for col in cat_cols:
        plt.figure(figsize=(6,3))
        df[col].value_counts().head(10).plot(kind="bar")
        plt.title(f"{name} — {col} (top 10)")
        plt.show()
```
